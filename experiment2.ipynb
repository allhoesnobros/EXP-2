Experiment 2: Study of Lists in Python

```python
list1=["apple", "orange", "cherry"]
print(list1)
```

```python
list2=[2, 2.5, "aayush", True]
print(list2)
```

```python
list3=["apple", "mango", "banana", "kiwi", "cherry", "orange"]
print(list3[2:5])
print(list3[:4])
print(list3[2:])
print(list3[2])
print(len(list3))
```

```python
list4= ["apple", "guava", "banana"]
list4[1]="mango"
print(list4)
```

```python
list5= ["apple", "guava", "banana"]
list5.append("orange")
print(list5)
list5.insert(1, "mango")
print(list5)
```

```python
list6= ["apple", "guava", "banana"]
list7= ["mango", "orange"]
list6.extend(list7)
print(list6)
```

```python
marks= [78, 85, 90, 66, 88]

print(max(marks))
print(min(marks))
print(sum(marks))
marks.sort()
print(marks)
marks.reverse()
print(marks)
print(len(marks))
print(sum(marks)/len(marks))
```

```python
GroceryList= ["eggs", "rice", "wheat", "chips", "diaper"]
ItemsPicked = "chips" # Define ItemsPicked for the first removal
GroceryList.remove(ItemsPicked)
print(GroceryList)
ItemsPicked = "diaper" # Redefine ItemsPicked for the second removal
GroceryList.remove(ItemsPicked)
print(GroceryList)
```
